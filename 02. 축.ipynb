# 2. 축

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # 윈도우인 경우
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Apple 인 경우
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False  # 한글 폰트 사용시, 마이너스 글자가 깨지는 현상을 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
plt.title('꺽은선 그래프', fontdict={'family':'Malgun Gothic', 'size':20})  # 개별 폰트 설정
```

```python
plt.plot(x, y)
plt.xlabel('X축')
plt.ylabel('Y축')
```

```python
plt.plot(x, y)
plt.xlabel('X축', color='red')
plt.ylabel('Y축', color='green')
```

```python
plt.plot(x, y)
plt.xlabel('X축', color='red', loc='right')  # left, center, right
plt.ylabel('Y축', color='green', loc='top')   #  top, center, bottom
```

```python
plt.plot(x, y)
plt.xticks([1, 2, 3])  # x좌표에 넣고 싶은 숫자만 적는다
plt.yticks([3, 6, 9, 12])
plt.show()
```
