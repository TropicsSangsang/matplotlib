```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # 윈도우인 경우
# matplotlib.rcParams['font.family'] = 'AppleGothic'  # Apple 인 경우
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False  # 한글 폰트 사용시, 마이너스 글자가 깨지는 현상을 해결
```

```python
labels = ['강백호', '서태웅', '정대만']  # 이름
values = [190, 187, 184]  # 키

plt.bar(labels, values)
```

```python
labels = ['강백호', '서태웅', '정대만']  # 이름
values = [190, 187, 184]  # 키

plt.barh(labels, values)  # h : horison
plt.xlim(175, 195)
```

```python
bar = plt.bar(labels, values)
bar[0].set_hatch('/')   # /////
bar[1].set_hatch('x')   # XXXXXX
bar[2].set_hatch('..')   # ......
```

```python
bar = plt.bar(labels, values)  # bar 에는 좌표상이므로 values 값이 저장

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height(), values[idx])
    # 그래프에 텍스트를 넣을건데, x축은 0,1,2 에 y축은 사각형의 높이에, values의 idx값을 넣어주면 될듯...
```

```python
bar = plt.bar(labels, values)  # bar 에는 좌표상이므로 values 값이 저장
plt.ylim(175, 195)
for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height() + 0.5, values[idx], ha='center', color='b')
```
